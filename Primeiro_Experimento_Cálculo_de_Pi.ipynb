##4.1. Encontrar a média das medidas


```
#dados experimentais coletados
dados = 3.197, 3.150, 3.147, 3.121, 3.142, 3.133, 3.154, 3.165, 3.116, 3.167
#definindo o cálculo da média
def average(values):
  total = 0
  nvalues = 0
  for value in values:
    total = total + value
    nvalues = nvalues + 1
  avg = total / nvalues
  return avg
  
def stdev(values):
  square_of_average = average(values)**2
  values_squared = []
  for value in values:
    values_squared.append(value**2)
  average_of_squares = average(values_squared)
  sd = (average_of_squares-square_of_average)**0.5
  return sd

#imprime o valor da média dos dados
print ("Média:", average(dados))

#imprime o valor do desvio padrão dos dados

print ("Desvio Padrão: ", stdev(dados))

#imprime o valor percentual dos dados
print ("Desvio Percentual: ", stdev(dados)/average(dados)*100)
```

##4.2 Gráficos
